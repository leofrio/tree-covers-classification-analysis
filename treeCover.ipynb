## ALTERING DATA

### making dataset simpler by transforming boolean into int value 

since soil_type is binary boolean and each row  can have only 1 soil type we can create a single column called soil_type aggregating the value to the type of soil it has, for example if a row has a 1.0 value on soil_type_23 , it certanly has 0 on all the other ones, so if we create a single column called soil_type and for this case add the value '23' we are indicating the value of soil we have thus reducing the complexity of the data base to the model

```python
soil_type_columns = df.filter(like='Soil_Type_')
soil_type_values = soil_type_columns.idxmax(axis=1)
# Extract the soil type number from the column name
df['soil_type'] = soil_type_values.apply(lambda x: int(x.split('_')[-1])) 
df=df.drop(columns=soil_type_columns) 
df
```

since wilderness_area is binary boolean and each row  can have only 1 wilderness_area we can create a single column called wilderness_area aggregating the value to the type of wilderness_area it has, for example if a row has a 1.0 value on wilderness_area_2 , it certanly has 0 on all the other ones, so if we create a single column called wilderness_area and for this case add the value 2 we are indicating the value of wilderness_area we have thus reducing the complexity of the data base to plotting

```python
wilderness_area_columns = df.filter(like='Wilderness_Area')
wilderness_area_values = wilderness_area_columns.idxmax(axis=1)
# Extract the soil type number from the column name
df['wilderness_area'] = wilderness_area_values.apply(lambda x: int(x.split('_')[-1])) 
df=df.drop(columns=wilderness_area_columns) 
df
```

#### applying dimensionality reduction  with tsne 

```python
from sklearn.manifold import TSNE 
print('splitting data')
df_tsne_sample=df.sample(frac=0.01) 
X_tsne_sample=df_tsne_sample.drop(columns=['Cover_Type']) 
y_tsne_sample=df_tsne_sample['Cover_Type']
print('initializing model')
tsne=TSNE(n_components=2,random_state=200)  #we apply the random state so that its not different everytime we restar 
print('Fitting data')
array_red=tsne.fit_transform(X_tsne_sample) #applying dimensional reductioon on array  
print('creating dataframe')
df_tsne=pd.DataFrame(array_red,columns=['COMP1','COMP2']) # making it into a pandas dataframe for easier viewing 
df_tsne = df_tsne.reset_index(drop=True)
y_tsne_sample = y_tsne_sample.reset_index(drop=True)
df_tsne
```

```python
import matplotlib.patches as mpatches
# Assuming df_elevation, df_height, and other necessary data are defined 
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27
df_elevation=df.copy()
conditions = [
    (df_elevation['Elevation'] < 2800),
    (df_elevation['Elevation'] >= 2800) & (df_elevation['Elevation'] <= 3100),
    (df_elevation['Elevation'] > 3100)
] 

choices = ['Less than 2800m', 'Between 2800m and 3100m', 'Over 3100m']
df_elevation['Elevation_Category'] = np.select(condlist=conditions, choicelist=choices, default='Error')

df_elevation['wilderness_area'] = pd.Categorical(df_elevation['wilderness_area'], categories=[0, 1, 2, 3], ordered=True)
df_elevation['Elevation_Category'] = pd.Categorical(df_elevation['Elevation_Category'], categories=['Less than 2800m', 'Between 2800m and 3100m', 'Over 3100m'], ordered=True)  

df_height = df_elevation.groupby('wilderness_area', observed=False).size().reset_index(name='count').copy()

palette = sns.color_palette('Set2', n_colors=8)
# Less than 2800m
lower = sns.barplot(x='wilderness_area', y='count', data=df_height, color=palette[2], edgecolor='none')  
for p in lower.patches:  
    height=p.get_height()     
    lower.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')  
# Between 2800m and 3100m
between_df = df_elevation[df_elevation['Elevation_Category'] == 'Between 2800m and 3100m'].groupby('wilderness_area', observed=False).size().reset_index(name='count')
middle = sns.barplot(x='wilderness_area', y='count', data=between_df, color=palette[3], edgecolor='none', bottom=[0, 0, 0, 0]) 

# Over 3100m
high_df = df_elevation[df_elevation['Elevation_Category'] == 'Over 3100m'].groupby('wilderness_area', observed=False).size().reset_index(name='count')
high = sns.barplot(x='wilderness_area', y='count', data=high_df, color=palette[4], edgecolor='none', bottom=between_df['count'])

# Calculate percentages
total_counts = df_height.groupby('wilderness_area')['count'].sum().reset_index()
merged_df = pd.merge(plot_df, total_counts, on='wilderness_area', suffixes=('_plot', '_total'))
plot_df['percentage'] = (merged_df['count_plot'] / merged_df['count_total']) * 100
j=0
# Annotations
for i in range(0,len(lower.patches),3): 
    lower_p=lower.patches[j]
    middle_p=middle.patches[j +4]
    high_p=high.patches[j+8] 
    j += 1 
    lower_height=plot_df.loc[i, 'count']
    middle_height=plot_df.loc[i+ 1, 'count']
    high_height=plot_df.loc[i+2, 'count'] 
    l_percentage = plot_df.loc[i, 'percentage']
    m_percentage = plot_df.loc[i+ 1, 'percentage']
    h_percentage = plot_df.loc[i+2, 'percentage']  
    p_height_l=0
    p_height_m=0
    p_height_h=0 
    modifier=0.3
    acceptable= 20 if df_height.loc[j-1,'count'] < 50000 else 10 if df_height.loc[j-1,'count'] < 100000 else 0
    if l_percentage > 0 and l_percentage >= acceptable:  
        p_height_l=middle_height +high_height + (lower_height *modifier)  
        lower.annotate(f'{int(lower_height)} ({l_percentage:.2f}%)', (lower_p.get_x() + lower_p.get_width() / 2., p_height_l ),
                        ha='center', va='center', xytext=(0, 10), textcoords='offset points',color='darkred',size='x-small')
    if m_percentage > 0 and m_percentage >= acceptable:  
        p_height_m=middle_height *modifier
        middle.annotate(f'{int(middle_height)} ({m_percentage:.2f}%)', (middle_p.get_x() + middle_p.get_width() / 2., p_height_m ),
                        ha='center', va='center', xytext=(0, 10), textcoords='offset points',color='darkgreen',size='x-small')
    if h_percentage > 0 and h_percentage >= acceptable: 
        p_height_h=middle_height + (high_height*modifier) 
        high.annotate(f'{int(high_height)} ({h_percentage:.2f}%)', (high_p.get_x() + high_p.get_width() / 2., p_height_h ),
                        ha='center', va='center', xytext=(0, 10), textcoords='offset points',color='darkblue',size='x-small')  

choices = ['Less than 2800m', 'Between 2800m and 3100m', 'Over 3100m']
bars = [mpatches.Patch(color=palette[i+2], label=choices[i]) for i in range(3)]
plt.legend(handles=bars, title='Elevation Category', bbox_to_anchor=(1.1, 1)) 

plt.ylim((0, 300000))
plt.ylabel('Amount')
plt.xlabel('Type of Wild Area')
plt.title('Number of Trees by Wild Area Type and Elevation')
plt.savefig('./images/n_tree_wilderness_area_and_elevation_stacked.png',)

plt.show()
```
